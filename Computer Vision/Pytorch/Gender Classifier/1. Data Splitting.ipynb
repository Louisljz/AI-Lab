## Divide Image into Folders

```python
import shutil
```

```python
male_path = os.path.join(directory, 'Male')
female_path = os.path.join(directory, 'Female')
```

```python
filenames = df_keys['File Name'].to_numpy()
keys = df_keys['Gender (M:1; F:0)'].to_numpy()
file_keys = dict(zip(filenames, keys))
```

```python
def move(gender_path, name):
  source_path = os.path.join(data_path, name)
  dest_path = os.path.join(gender_path, name)
  shutil.move(source_path, dest_path)
```

```python
for name in os.listdir(data_path):
  if file_keys[name] == 1:
    move(male_path, name)
  else:
    move(female_path, name)
```
