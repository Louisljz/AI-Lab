## Remove Stopwords

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

# Importing stopwords list from Sastrawi
factory = StopWordRemoverFactory()
stop_words = factory.get_stop_words()
print(stop_words)
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('indonesian'))
print(stop_words)
```

```python
print(word_tokens)
```

```python
filtered_words = []
for token in word_tokens:
  if token.lower() not in list(stop_words):
    filtered_words.append(token.lower())

print(filtered_words)
```

## Stemming

```python
from nltk.stem import PorterStemmer
```

```python
porter_stemmer  = PorterStemmer()
sentence = 'studies studying cries cry'
tokenization = nltk.word_tokenize(sentence)

for word in tokenization:
    print("Stemming for {} is {}".format(word, porter_stemmer.stem(word)))  
```

## Lemmatization

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer

wordnet_lemmatizer = WordNetLemmatizer()
sentence = 'studies studying cries cry'
tokenization = nltk.word_tokenize(sentence)
for word in tokenization:
    print("Lemma for {} is {}".format(word, wordnet_lemmatizer.lemmatize(word)))
```
