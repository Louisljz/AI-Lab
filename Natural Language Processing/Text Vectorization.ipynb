## Mount Drive and Set Up Paths

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
```

## 1. Train Model Using Gensim

```python
# define training data
sentences = [['this', 'is', 'the', 'first', 'sentence', 'for', 'word2vec'],
             ['this', 'is', 'the', 'second', 'sentence'],
             ['yet', 'another', 'sentence'],
             ['one', 'more', 'sentence'],
             ['and', 'the', 'final', 'sentence']]
```

```python
# train model
word2vec_model = Word2Vec(sentences, size=80, min_count=1)
print(word2vec_model)
```

```python
# save model
word2vec_model.save('word2vec_model.bin')
```

## 2. Use Pre-trained Model

```python
glove_input_file = 'glove.6B.100d.txt'
word2vec_output_file = 'glove.6B.100d.word2vec'
```

```python
glove2word2vec(glove_input_file, word2vec_output_file)
```

```python
model = KeyedVectors.load_word2vec_format(word2vec_output_file, binary=False)
```

```python
result = model.most_similar(positive=['fast', 'low'], negative=['high'], topn=1)
```
